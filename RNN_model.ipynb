```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

model = Sequential([
    SimpleRNN(10, input_shape=(5, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

model.summary()
```
