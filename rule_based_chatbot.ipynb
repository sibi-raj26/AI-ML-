```python
print("Bot: Welcome! Type 'bye' to exit.")

responses = {
    "hello": "Hi! How can I help you?",
    "how are you": "I am doing well.",
    "what is your name": "I am a Rule-Based Chatbot.",
    "python": "Python is a popular programming language.",
    "java": "Java is an object-oriented programming language."
}

while True:

    user = input("You: ").lower()

    if user == "bye":
        print("Bot: Goodbye!")
        break

    elif user in responses:
        print("Bot:", responses[user])

    else:
        print("Bot: Sorry, I don't understand that question.")
```
